```python
from dotenv import load_dotenv
load_dotenv()

import getpass
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

if "GOOGLE_API_KEY" not in os.environ :
  os.environ["GOOGLE_API_KEY"] =  getpass.getpass("Enter you google API :")
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
template =ChatPromptTemplate.from_template("tell me a fun fact about {topic}.")
parser = StrOutputParser()
```

```python
prompt_value = template.invoke({"topic" : "penguin"})
response_obj = llm.invoke(prompt_value)
final_text = parser.invoke(response_obj)
print(final_text)
```

```python
prompt_value = template.invoke({"topic" : "arijit singh"})
response_obj = llm.invoke(prompt_value)
final_text = parser.invoke(response_obj)
print(final_text)
```

```python
chain = template | llm | parser
chain.invoke({"topic":"penguin"})
```

```python
# Install required packages
# !pip install -U langchain langchain-core langchain-google-genai python-dotenv

# Load environment variables
from dotenv import load_dotenv
load_dotenv()

import os
import getpass

# Ask for API key if not already set
if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Google API Key: ")

# Imports
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Initialize LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.0
)

# One-line LCEL chain
chain = ChatPromptTemplate.from_template(
    "Given the movie '{movie}', state its release year and calculate how many years ago it was released from 2026."
) | llm | StrOutputParser()

# Invoke
print(chain.invoke({"movie": "Mission Mangal"}))
```
